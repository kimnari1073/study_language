## Lab 7:파생변수 추가하기
df.assign(컬럼이름=함수)

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
#math, english, science 값을 합산한 값을 갖는 파생변수 total 생성
exam.assign(total = exam['math']+exam['english']+exam['science']).head()
#기존 데이터 프레임에는 추가되지 않는다.
```

```python
import numpy as np
#science가 60점 초과면 pass, 60점 이하면 fail값을 갖는 파생변수 test추가
exam.assign(test = np.where(exam['science']>60,'pass','fail')).head()
```

## Lab 7(혼자 해보기): mpg 데이터를 이용하여 문제 분석

Q1:mpg 사본을 만들고, cty와 hwy를 더한 '합산 연비 변수'total을 추가하시오

```python
import pandas as pd
mpg_cp=pd.read_csv('mpg.csv')
mpg_cp.head()
```

```python
mpg_cp['total']=mpg_cp['cty']+mpg_cp['hwy']
mpg_cp.head()
```

Q2:합산연비 total을 2로 나누어 평균 연비 변수(mean)를 추가하시오

```python
mpg_cp['mean'] = mpg_cp['total']/2
mpg_cp.head()
```

Q3:평균 연비가 가장 높은 자동차 3종의 데이터를 출력하시오

```python
mpg_cp.sort_values('mean',ascending=False).head(3)
```
