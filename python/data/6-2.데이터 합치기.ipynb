## 데이터 합치기 (merge(), concat())

### Lab 6:가로로 합치기 merge()

```python
import pandas as pd
# 중간고사 점수를 갖는 데이터 프레임 임의 생성
test1 = pd.DataFrame({'id':[1,2,3,4,5,6],
                     'midterm':[60,80,70,90,85,66]})
test1
```

```python
#기말고사 점수를 갖는 데이터 프레임 임의 생성
test2 = pd.DataFrame({'id':[1,2,3,4,5,7],
                     'finalterm':[70,83,65,95,80,77]})
test2
```

```python
#데이터 프레임test1과 test2를 가로로 합치기 (how='left'), default는 inner
total = pd.merge(test1,test2,how='left',on='id')
total
```

### Lab 7:세로로 합치기 concat()

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})
group_a
```

```python
# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
group_b
```

```python
# 데이터 합쳐서 group_all에 할당
group_all = pd.concat([group_a, group_b])
group_all
```
